<h4>1.1.3 Basketball Position</h4>
<p>
In the sport of basketball, there are five players that play per team, each assigned to positions. Historically, these players have been assigned to positions defined by the role they play on the court, from a strategic point of view. The three main positions are guard, forward, and center, with the standard team featuring two guards, two forwards, and a center. Over time, as more specialized roles developed, each of the guards and forwards came to be differentiated, and today each of the five positions are known by unique names, each of which has also been assigned a number: point guard (PG), the shooting guard (SG), the small forward (SF), the power forward (PF), and the center (C).<sub>[1]</sub>

Point guards are expected to run the team's offense by controlling the ball and making sure that it gets to the right player at the right time. They must also be able to adapt to what the defense is allowing and must control the pace of the game. Their primary job is to facilitate scoring opportunities for their team, or sometimes for themselves.<sub>[2]</sub>
    
Shooting guards are expected to be a threat to his opponent as he’s able to score from anywhere on the court. Typically, they are stronger, taller, more athletic than point guard and good in long range shots. Their primary job is to score points.<sub>[3]</sub>
    
Small forward is considered to be perhaps the most versatile of the five main basketball positions as they contribute offensively and defensively. They are responsible for scoring points and defending, and often are secondary or tertiary rebounders behind the power forward and center.<sub>[4]</sub>
    
Power forward is often the team's most powerful and dependable scorer, being able to score close to the basket while also being able to shoot mid-range jump shots. They are also very crafty and have to be versatile on both offense and defense but not as much as a small forward.<sub>[5]</sub>
    
Center is normally the tallest player on the team, and often has a great deal of strength and body mass as well. They are valued for their ability to protect their own goal from high-percentage close attempts on defense, while scoring and rebounding with high efficiency on offense.<sub>[6]</sub>

<h4>3.2.3 Gather all teams' URL</h4>
<img src="images/bb_team_url.png">
<p>
The above image shows how basketball-reference home page looks like, you can notice that in the second section "Every NBA Team", there is an option tag that will allow user to select which team to view. Thus, every option contains the value of their website url respectively.

In order to make our life easier, we can make use of it and webscrape down all the value in the option so that we will just need to combine two string base_url and the option's value together to browse to the target team websites.
    
Now, we can use the beautifulsoup.find() to locate the target html tag with the attribute "id: teams" since the option tags are inside the div with id called teams. After that, we can use beautifulsoup.find_all() method to target all option tags inside the div with id called teams and run a loop to get their value out and save it into a list called team_url_list
</p>

```python
# find() method searches for a tag and specified attributes, returning the first match
select_team = base_soup.find(name = 'div', attrs = {'id' : 'teams'})

# Creating a list of url endings for each NBA team's webpage
team_url_list = []
for i in select_team.find_all('option')[1:31]:
    name = i.get('value')
    team_url_list.append(name + "/2022.html")
team_url_list.sort() # sort list by a-z
team_url_list
```

<h4>3.2.4 Gather all teams' abbreviation</h4>
<p>
After getting the all team url, we will need to get their abbreviation too because some of the team's name are too lengthy! We can do it by extracting the abbreviation from the team_url_list.
    
for example:
/teams/ATL/2022.html this is first item that stored in the team_url_list. What we need is the string between /teams/ and /2022.html.
    
In this case, we can use the regular expression library to help us out. First, we need to use search() method and pass in the delimiter parameter. The syntax (.*) stands for what our target is which is the abbreviation this time, and the two string before and after it will acknowledge the library when to start and when to stop for targeting and then the result will be save to a variable called team.
    
Then, we can use RegEx.group().strip() to get the abbreviation and store it into team_abbreviation_list for later use.  
</p>

```python
team_abbreviation_list = []

# traverse through the list
for url_link in team_url_list:
    team = re.search(r'/teams/(.*)/2022.html',url_link)
    team_abbreviation_list.append(team.group(1).strip())
pd.Series(team_abbreviation_list)
```

<h4>3.2.5 Scrape Team Data</h4>
<p>
Next we will attempt to scrape all 30 NBA teams stats. If successful, we will append the scraped information to the list called team_standings.
    
First we will do something similar previously, store the team standings url into a variable target_url. Then, we will requests the server to get a copy of its HTML document and then we will parse it with beautifulsoup to allow us to work on it later on.
</p>

<img src="images/team_standings.png">

<p>
You may notice the code below why there are two seperate loop that do the same task. This is because we have two different table for the teams that is East and West of North America. Thus, instead of doing it with if condition I decided to use two loop as this will make the code more readable.
    
Besides that, you may also notice why we start from table row one not zero, this is because the zeroth row is the header which is not the information we wanted to scrape. This approach will be used to scrape other information later on too!
</p>

<p>
We now know that there are two table that have the information we needed. Therefore, we will use beautifulsoup.find() method to locate first table which is the table with id called confs_standings_E and use a loop that start from one till the last table row. In additional, we will use beautifulsoup.find() to search the current row in the loop and extract the data out. All the scraped data will be store into a dictionary that will have key and value respectively. Then, we will store each team dictionary into a list called team_standings.
</p>

```python
team_standings = [] # store every team stats in 2021-22 season

target_url = 'https://www.basketball-reference.com/leagues/NBA_2022_standings.html'

# send a GET request to the target_url
res_target = requests.get(target_url)

# parses the content of an HTML document
target_soup = BeautifulSoup(res_target.content, 'lxml')

# Eastern Conference Teams
target_team_standings = target_soup.find(name = 'table', attrs = {'id' : 'confs_standings_E'})

for row in target_team_standings.find_all('tr')[1:]:  # excluding the first 'tr', since that's the table's title head
    team = {}
    team['Name'] = row.find('a').text.strip()
    team['Win'] = row.find('td', {'data-stat' : 'wins'}).text
    team['Lose'] = row.find('td', {'data-stat' : 'losses'}).text
    team['Win/Loss %'] = row.find('td', {'data-stat' : 'win_loss_pct'}).text
    team['Game Behind'] = row.find('td', {'data-stat' : 'gb'}).text
    team['Pts Per Game'] = row.find('td', {'data-stat' : 'pts_per_g'}).text
    team['Opp Pts Per Game'] = row.find('td', {'data-stat' : 'opp_pts_per_g'}).text
    team['Simple Rating System'] = row.find('td', {'data-stat' : 'srs'}).text
    team_standings.append(team)

# Western Conference Teams
target_team_standings = target_soup.find(name = 'table', attrs = {'id' : 'confs_standings_W'})

for row in target_team_standings.find_all('tr')[1:]:  # excluding the first 'tr', since that's the table's title head
    team = {}
    team['Name'] = row.find('a').text.strip()
    team['Win'] = row.find('td', {'data-stat' : 'wins'}).text
    team['Lose'] = row.find('td', {'data-stat' : 'losses'}).text
    team['Win/Loss %'] = row.find('td', {'data-stat' : 'win_loss_pct'}).text
    team['Game Behind'] = row.find('td', {'data-stat' : 'gb'}).text
    team['Pts Per Game'] = row.find('td', {'data-stat' : 'pts_per_g'}).text
    team['Opp Pts Per Game'] = row.find('td', {'data-stat' : 'opp_pts_per_g'}).text
    team['Simple Rating System'] = row.find('td', {'data-stat' : 'srs'}).text
    team_standings.append(team)
```

<img src="images/player_stats.png">
<p>
For narrowing my search to per game statistics, I will use beautifulsoup’s.find() method to target the ‘table’ tag, along with an id attribute of ‘per_game’. After the appropriate tag and attribute were found by searching through the webpage’s HTML and analyzing the text that highlighted my table of interest, as shown below. The isolated ‘Per Game’ table is now named target_per_game and this will be iterate repeatedly for every team in order to get all 30 team's player data.
    
Then, I can now go through each row (player) in that table and extract the data and save it into a dictionary called player. Besides that, I will also browse to the every individual player page to extract their height, weight and position as these are not included in the target_per_game. I will store the player bio into a variable called player_info with beautifulsoup and find out all the p tags. After that, RegEx will be used here again to get the height and weight from the p tags. Furthermore, the player bio will also be saved into the player (dictionary).
    
In brief, all player that are in the same team will be saved into a list called target_stats and all team (target_stats) will be saved into player_stats.
</p>

```python
# version 1.0
player_stats = [] # store all player stats in 2021-22 season

count = 0 # keep track of how many requests made

for team_url in team_url_list[0:]:
    # combine the base and target into each team link
    target_url = base_url + team_url

    # send a GET request to the target_url
    res_target = requests.get(target_url)

    count += 1; # increase by 1 every request made

    # parses the content of an HTML document
    target_soup = BeautifulSoup(res_target.content, 'lxml')

    target_per_game = target_soup.find(name = 'table', attrs = {'id' : 'per_game'})

    target_stats = [] # store each player stats on that team

    for row in target_per_game.find_all('tr')[1:]:  # excluding the first 'tr', since that's the table's title head
        player = {}
        player['Name'] = row.find('a').text.strip()
        player['Age'] = row.find('td', {'data-stat' : 'age'}).text
        player['Min PG'] = row.find('td', {'data-stat' : 'mp_per_g'}).text
        player['Field Goal %'] = row.find('td', {'data-stat' : 'fg_pct'}).text
        player['Rebounds PG'] = row.find('td', {'data-stat' : 'trb_per_g'}).text
        player['Assists PG'] = row.find('td', {'data-stat' : 'ast_per_g'}).text
        player['Steals PG'] = row.find('td', {'data-stat' : 'stl_per_g'}).text
        player['Blocks PG'] = row.find('td', {'data-stat' : 'blk_per_g'}).text
        player['Turnovers PG'] = row.find('td', {'data-stat' : 'tov_per_g'}).text
        player['Points PG'] = row.find('td', {'data-stat' : 'pts_per_g'}).text

        player_url = ('https://www.basketball-reference.com/' + row.find('a').attrs['href'])
        player_rest = requests.get(player_url) # send a GET request to get player's HTML
        player_soup = BeautifulSoup(player_rest.content, 'lxml') # parses the content of an HTML document

        if len(player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')) > 1:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[1]
        else:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[0]

        count += 1 # increase by 1 every request made

        s = str(player_info.find_all('p'))

        # regular expression session
        weight = re.search(r'\,[\s]<span>(.*)lb</span>', s)
        position = re.search('Position:\n  </strong>\n  (.*)\n\n', s)
        height = re.search('<p><span>(.*)</span>\,', s)
        player['Height'] = height.group(1).strip()
        player['Weight (Lbs)'] = weight.group(1).strip()
        player['Position'] = position.group(1).strip()

        target_stats.append(player) # push that player to his team

        print("length: " + str(len(player_stats)) + ", count: " + str(count))

    player_stats.append(target_stats) # push that team into the list

    print("length: " + str(len(player_stats)) + ", count: " + str(count))
```

```python
# version 1.1
player_stats = [] # store all player stats in 2021-22 season

count = 0 # keep track of how many requests made

for team_url in team_url_list[0:]:
    # combine the base and target into each team link
    target_url = base_url + team_url

    # send a GET request to the target_url
    res_target = requests.get(target_url)

    count += 1; # increase by 1 every request made

    # parses the content of an HTML document
    target_soup = BeautifulSoup(res_target.content, 'lxml')

    target_per_game = target_soup.find(name = 'table', attrs = {'id' : 'per_game'})

    target_stats = [] # store each player stats on that team

    for row in target_per_game.find_all('tr')[1:]:  # excluding the first 'tr', since that's the table's title head
        player = {}
        player['Name'] = row.find('a').text.strip()
        player['Age'] = row.find('td', {'data-stat' : 'age'}).text
        player['Min PG'] = row.find('td', {'data-stat' : 'mp_per_g'}).text
        player['Field Goal %'] = row.find('td', {'data-stat' : 'fg_pct'}).text
        player['Rebounds PG'] = row.find('td', {'data-stat' : 'trb_per_g'}).text
        player['Assists PG'] = row.find('td', {'data-stat' : 'ast_per_g'}).text
        player['Steals PG'] = row.find('td', {'data-stat' : 'stl_per_g'}).text
        player['Blocks PG'] = row.find('td', {'data-stat' : 'blk_per_g'}).text
        player['Turnovers PG'] = row.find('td', {'data-stat' : 'tov_per_g'}).text
        player['Points PG'] = row.find('td', {'data-stat' : 'pts_per_g'}).text

        player_url = ('https://www.basketball-reference.com/' + row.find('a').attrs['href'])
        player_rest = requests.get(player_url) # send a GET request to get player's HTML
        player_soup = BeautifulSoup(player_rest.content, 'lxml') # parses the content of an HTML document

        if len(player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')) > 1:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[1]
        else:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[0]

        count += 1 # increase by 1 every request made

        s = str(player_info.find_all('p'))

        # regular expression session
        weight = re.search(r'\,[\s]<span>(.*)lb</span>', s)
        position = re.search('Position:\n  </strong>\n  (.*)\n\n', s)
        height = re.search('<p><span>(.*)</span>\,', s)
        player['Height'] = height.group(1).strip()
        player['Weight (Lbs)'] = weight.group(1).strip()
        player['Position'] = position.group(1).strip()

        target_stats.append(player) # push that player to his team

        time.sleep(3.5) # wait for 3.5 seconds

    player_stats.append(target_stats) # push that team into the list

    print("team count: " + str(len(player_stats)) + ", player count: " + str(count))
    time.sleep(3.5) # wait for 3.5 seconds

show_df = pd.DataFrame(player_stats)
```

Therefore, when accessing the player we need to go through the team first then the index of that player before know any of his stats which is considered quite complex in my opinion.

In contrast, we will try to make our life easier with new data structure which is using a list(array) to store all stats of that player first and then push it into another list called player_stats_list. By doing so, we will get a crystal clear result that show the stats directly without accessing too many layer.

In additional, I added the player's team into account which still allow us to keep track of the player's team because we are now storing all player together without categorising them into team unlike previous data structure that keep all player in the same team in one inner list and then it will be store into another outer list.

```python
# version 2.0
player_stats_list = [] # store all player stats in 2021-22 season

count = 0 # keep track of how many requests made
index = 0 # keep track of how many team completed

for team_url in team_url_list[0:]:
    # combine the base and target into each team link
    target_url = base_url + team_url

    # send a GET request to the target_url
    res_target = requests.get(target_url)

    count += 1; # increase by 1 every request made

    # parses the content of an HTML document
    target_soup = BeautifulSoup(res_target.content, 'lxml')

    target_per_game = target_soup.find(name = 'table', attrs = {'id' : 'per_game'})

    # target_stats = [] # store each player stats on that team

    for row in target_per_game.find_all('tr')[1:]:  # excluding the first 'tr', since that's the table's title head
        player = []
        player.append(team_abbreviation_list[index]) # Team
        player.append(row.find('a').text.strip()) # Name
        player.append(row.find('td', {'data-stat' : 'age'}).text) # Age
        player.append(row.find('td', {'data-stat' : 'mp_per_g'}).text) # Minutes per Game
        player.append(row.find('td', {'data-stat' : 'fg_pct'}).text) # Field Goal %
        player.append(row.find('td', {'data-stat' : 'trb_per_g'}).text) # Rebounds per Game
        player.append(row.find('td', {'data-stat' : 'ast_per_g'}).text) # Assits per Game
        player.append(row.find('td', {'data-stat' : 'stl_per_g'}).text) # Steals per Game
        player.append(row.find('td', {'data-stat' : 'blk_per_g'}).text) # Blocks per Game
        player.append(row.find('td', {'data-stat' : 'tov_per_g'}).text) # Turnovers per Game
        player.append(row.find('td', {'data-stat' : 'pts_per_g'}).text) # Points per Game

        player_url = ('https://www.basketball-reference.com/' + row.find('a').attrs['href'])
        player_rest = requests.get(player_url) # send a GET request to get player's HTML
        player_soup = BeautifulSoup(player_rest.content, 'lxml') # parses the content of an HTML document

        if len(player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')) > 1:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[1]
        else:
            player_info = player_soup.find(name = 'div', attrs = {'id' : 'meta'}).find_all('div')[0]

        count += 1 # increase by 1 every request made

        s = str(player_info.find_all('p'))

        # regular expression session
        weight = re.search(r'\,[\s]<span>(.*)lb</span>', s)
        position = re.search('Position:\n  </strong>\n  (.*)\n\n', s)
        height = re.search('<p><span>(.*)</span>\,', s)
        player.append(height.group(1).strip())
        player.append(weight.group(1).strip())
        player.append(position.group(1).strip())

        player_stats_list.append(player)

        time.sleep(3.5) # wait for 3.5 seconds

    index += 1

    print("team count: " + str(index) + ", player count: " + str(count))
    time.sleep(3.5) # wait for 3.5 seconds

show_df = pd.DataFrame(player_stats_list)
```

<h3>5.2 Tokenise the Position</h3>
<p>
The function will seperate the input with delimiters and return a list of string. It takes a string parameter and then will split it according to ',' or 'and'.
</p>
<p>
First version, I tried to use python standard library to tokenise the input string. The str.split() will allow us to set our delimiter and seperate accordingly. However, it fails to detect the 'and' and seperate it from the string.
</p>

```python
# version 1.0
def tokenise(position):
    return position.split(',' or ' and' or 'and ')

tokenise('Small Forward, Power Forward, Point Guard, and Shooting Guard')

# remark: feel free to comment out the code below to see how my function react!

# tokenise('Small Forward and Shooting Guard')
# tokenise('Small Forward)
```

Second version, I use the regular expression library which will have more powerful feature to help us with the task. The re.split() allows us to seperate the input string accord to the delimiters and return a list of string that had been tokenised. Fortunely, it managed to recognise the 'and' word and produce the result that I wanted. However, there is still flaw where you will notice below that there is a empty string in the return list.

```python
# version 2.0
def tokenise(position):
    return re.split(', |and | and ', position) # regular expression

tokenise('Small Forward, Power Forward, Point Guard, and Shooting Guard')

# remark: feel free to comment out the code below to see how my function react!

# tokenise('Small Forward and Shooting Guard')
# tokenise('Small Forward)
```

Third version, I use the combination of if condition and list comprehension to traverse the list and remove any empty string. In addition, you can now pass the delimiter as a parameter which makes the function more versatile.

```python
# version 3.0
def tokenise(position, delimiters):
    temp = re.split(delimiters, position) # regular expression
    return [i for i in temp if i != ''] # check and remove the empty string

tokenise('Small Forward, Power Forward, Point Guard, and Shooting Guard', ', |and | and ')

# remark: feel free to comment out the code below to see how my function react!

# tokenise('Small Forward and Shooting Guard')
# tokenise('Small Forward)
```

<h3>5.3 Back to the main route</h3>
<p>
After solving the issue above, we can now move one step close to reveal who is our MVP. Now, we need to use the tokenise function to help us to split the position and take whichever first two into calculation.
    
Then, we will use my own-defined algorithm to grade each player according to their position and stats. You must be wondering where do I get this algorithm from... actually it's based on my personal opinion who has been a fan of NBA for more than seven years! Therefore, the output might not be approved by other people but this is how I think the role of each player and give the weight of each stats accordingly.

Position Weightage by ME:
<ul>
    <li>Point Guard: 30% points, 30% assists, 15% steals, 15% rebounds, 10% blocks</li>
    <li>Shooting Guard: 40% points, 20% assists, 15% steals, 15% rebounds, 10% blocks</li>
    <li>Small Forward: 30% points, 20% assists, 20% steals, 20% rebounds, 10% blocks</li>
    <li>Power Forward: 30% points, 10% assists, 10% steals, 25% rebounds, 25% blocks</li>
    <li>Center: 30% points, 5% assists, 5% steals, 30% rebounds, 30% blocks</li>
</ul>

player that has single position,
<br>
    total marks = sum (each stats * weightage) - turnovers

player that has multiple position,
<br>
    total marks = [sum (each stats * weightage) - turnovers] / iterate
</p>

```python
def getWeightage(temp_position):
    if temp_position == 'Point Guard':
        return [0.3, 0.3, 0.15, 0.15, 0.1]
    elif temp_position == 'Shooting Guard':
        return [0.4, 0.2, 0.15, 0.15, 0.1]
    elif temp_position == 'Small Forward':
        return [0.3, 0.2, 0.2, 0.2, 0.1]
    elif temp_position == 'Power Forward':
        return [0.3, 0.1, 0.1, 0.25, 0.25]
    elif temp_position == 'Center':
        return [0.3, 0.05, 0.05, 0.3, 0.3]

def getOverallMark(player, weightage):
    return sum([
        float(player['Points PG']) * weightage[0],
        float(player['Assists PG']) * weightage[1],
        float(player['Steals PG']) * weightage[2],
        float(player['Rebounds PG']) * weightage[3],
        float(player['Blocks PG']) * weightage[4]
    ]) - float(player['Turnovers PG'])
```

<h3>5.4 One more step to our MVP!</h3>
<p>
After defining the function above, we can now calculate each player mark based on our algorithm and see who is our MVP!
</p>

<p>
We will have an outer for loop that traverse through all teams, followed by another inner loop that will traverse all players in that team. Then, we will tokenise each player's position and use it to check how many iterations needed in order to calculate the mark.
</p>

<p>
If the player has one position then we will calculate each stats according to the weightage and sum them up and then deduct by turnover. If the player has more than one position then we will only take the first two position into account by calculate each stats according to the weightage and sum them up and then deduct by turnover twice repectively and then divide by number of iteration, in this case will be two at most.
</p>

<p>
Next, we will save the player's name as a key and the mark as a value into the result dictationary which will then be pushed into player_mark list.
</p>

```python
player_mark = []

# traverse through every player
for row in range(len(player_stats_list_df)):
    player = player_stats_list_df.loc[row]

    # check if there is any None to prevent error
    if player is not None:
        result = {}
        mark = 0

        # tokenise the position and store into variable
        position = tokenise(player['Position'], ', |and | and ')

        # check how many calculate we need for that player
        # because some might have more than one position
        # we will only take the first two position into account
        if len(position) >= 2:
            iterate = 2
        else:
            iterate = 1

        # calculating the mark
        for i in range(iterate):
            # get weightage for that position and store list into variable
            weightage = getWeightage(position[i])
            # calculate the mark according to player's stats and his position
            mark += getOverallMark(player, weightage)
        # save the output as dict
        result['Name'] = player['Name']
        result['Mark'] = mark / iterate
        # push it into list
        player_mark.append(result)
```

</p>

```python
player_mark_df = pd.DataFrame(player_mark)
player_mark_df.describe()
```
